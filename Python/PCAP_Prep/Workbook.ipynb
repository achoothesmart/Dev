```python
class Stack:
    def __init__(self):
        self.__stk = []

    def push(self, val):
        self.__stk.append(val)

    def pop(self):
        val = self.__stk[-1]
        del self.__stk[-1]
        return val


class CountingStack(Stack):
    def __init__(self):
        Stack.__init__(self)
        self.__sum = 0

    def get_counter(self):
        return self.__sum

    def pop(self):
        val = Stack.pop(self)
        self.__sum += val
        return val
	

stk = CountingStack()
for i in range(10):
    stk.push(i)
    stk.pop()
print(stk.get_counter())

print(stk.__dict__)
```

```python
class IterEvens:
    def __init__(self, n):
        self.__n = n
        self.__i = 0

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.__i >= self.__n:
            raise StopIteration
        else:
            self.__i += 1
            val = self.__i * 2
            return val
            
for i in IterEvens(5):
    print(i, end=" ") # 2 4 6 8 10
```

```python
def powers_of_2(n):
    for i in range(n):
        yield 2 ** i

print('Using powers_of_2 as a generator:')
for i in powers_of_2(5):
    print(i, end=" ") # 1 2 4 8 16

print('\n\nUsing powers_of_2 as a filter:')
for i in range(20):
    if i in powers_of_2(20):
        print(i, end=" ")
```

```python
print('Using powers_of_2 as a generator expression:')
for x in (el * 2 for el in range(5)):
    print(x, end=" ") # 0 2 4 6 8

print('\n\nUsing powers_of_2 as a list comprehension:')

for x in [el * 2 for el in range(5)]:
    print(x, end=" ") # 0 2 4 6 8
```
