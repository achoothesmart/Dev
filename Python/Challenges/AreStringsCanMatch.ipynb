```python
s1 = 'bank'
s2 = 'kanb'

def findIfOneSwapCanMatchStrings(s1: str, s2: str):
    indexs = []

    if len(s1) != len(s2):
        return False
    
    
    for i in range(0, len(s1)):
        if(s1[i] != s2[i]):
            indexs.append(i)
        if(len(indexs) > 2):
            return False
    
    i1 = indexs[0]
    i2 = indexs[1]
    if( s1[i1] == s2[i2] and s1[i2] == s2[i1]):
        return True
    
    return False


res = findIfOneSwapCanMatchStrings(s1, s2)
if(res):
    print('its a Match')
else:
    print('Ooooo.. its a waste of time!!')
```
