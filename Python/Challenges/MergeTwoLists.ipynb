```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
def mergeTwoLists(list1: ListNode, list2: ListNode) -> list:
    i=j=0
    list3 = []
    stop=False
    while(not stop):
        if(len(list1) > i and len(list2) > j):
            if(list1[i] < list2[j]):
                list3.append(list1[i])
                i += 1
            else:
                list3.append(list2[j])
                j += 1
        else:
            if(i == len(list1)):
                # i reached end, so copy all remaining list2
                list3.extend(list2[j:])
                break
            elif(j == len(list2)):
                # j reached end, so copy all remaining list1
                list3.extend(list1[i:])
                break
    return list3
```

```python
list1 = [1,2,4]
list2 = [1,3,4]

list3 = mergeTwoLists(list1, list2)
print(list3)
```
